```python
import tensorflow as tf
import numpy as np
```

```python
tf.enable_eager_execution()
tf.__version__
```

```python
X= np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
], dtype="float32")

y= np.array([
    [0],
    [0],
    [0],
    [1]
], dtype="float32")
```

```python
X
```

```python
y
```

```python
w= tf.Variable([
    [1],
    [2]
], dtype="float32")
b= tf.Variable([3], dtype="float32")
```

```python
w
```

```python
b
```

```python
tf.matmul(X,w) # X * w
```

```python
tf.matmul(X,w)+b # X * w + b
```

```python
tf.sigmoid(tf.matmul(X,w)+b)
```

```python
optimizer = tf.train.AdamOptimizer(0.1) # w, b 값을 자동으로 수정해줌
```

```python
with tf.GradientTape() as tape:
  hypothesis = tf.sigmoid(tf.matmul(X,w)+b)
  print("hypothesis=", hypothesis)
  cost = -tf.reduce_mean(y * tf.log(hypothesis) + (1 - y) * tf.log(1 - hypothesis))
  print("cost=", cost)
  # reduce_mean : sum과 같은 역할
```

```python
with tf.GradientTape() as tape:
 
  hypothesis = tf.sigmoid(tf.matmul(X,w)+b)
  print("hypothesis=", hypothesis)
  
  cost = -tf.reduce_mean(y * tf.log(hypothesis) + (1 - y) * tf.log(1 - hypothesis))
  print("cost=", cost)
  # reduce_mean : sum과 같은 역할
  
  grads=tape.gradient(cost,[w,b])
  print("기울기 grads=", grads)

  print("w 업데이트 전=", w)
  print("b 업데이트 전=", b)

  print("zip(grads,[w,b])=",list(zip(grads,[w,b])) )
  optimizer.apply_gradients(grads_and_vars=zip(grads,[w,b]))
  print("w 업데이트 후=", w)
  print("b 업데이트 후=", b)
```
