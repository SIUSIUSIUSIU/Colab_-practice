```python
hypothesis0 = tf.sigmoid(tf.matmul(X_train,w0)+b0) # 첫번째 회귀
hypothesis1 = tf.nn.softmax(tf.matmul(hypothesis0,w1)+b1) # 두번째 회귀
print("hypothesis1=",hypothesis1[0:3])
print("y_train=",y_train[0:3])
cost = -tf.reduce_sum(y_train * tf.log(hypothesis1)) / len(hypothesis1)
print("cost=",cost) 
```

```python
with tf.GradientTape() as tape: 
  hypothesis0 = tf.sigmoid(tf.matmul(X_train,w0)+b0) # 첫번째 회귀
  hypothesis1 = tf.nn.softmax(tf.matmul(hypothesis0,w1)+b1) # 두번째 회귀
  print("hypothesis1=",hypothesis1[0:3])
  print("y_train=",y_train[0:3])
  cost = -tf.reduce_sum(y_train * tf.log(hypothesis1)) / len(hypothesis1)
  print("cost=",cost) 
  grads = tape.gradient(cost, [w0,w1,b0,b1])
  print("grads=",grads)
```

```python
optimizer = tf.train.AdamOptimizer(0.1)
```

```python
with tf.GradientTape() as tape: 
  hypothesis0 = tf.sigmoid(tf.matmul(X_train,w0)+b0) # 첫번째 회귀
  hypothesis1 = tf.nn.softmax(tf.matmul(hypothesis0,w1)+b1) # 두번째 회귀
  print("hypothesis1=",hypothesis1[0:3])
  print("y_train=",y_train[0:3])
  cost = -tf.reduce_sum(y_train * tf.log(hypothesis1)) / len(hypothesis1)
  print("cost=",cost) 
  grads = tape.gradient(cost, [w0,w1,b0,b1])
  # print("grads=",grads)
  print("w0 업제이트 전=",w0)
  print("w1 업제이트 전=",w1)
  print("b0 업제이트 전=",b0)
  print("b1 업제이트 전=",b1)
  optimizer.apply_gradients(grads_and_vars=zip(grads,[w0,w1,b0,b1]))
  print("w0 업제이트 후=",w0)
  print("w1 업제이트 후=",w1)
  print("b0 업제이트 후=",b0)
  print("b1 업제이트 후=",b1)
```

```python
for step in range(100):
  with tf.GradientTape() as tape: 
    hypothesis0 = tf.sigmoid(tf.matmul(X_train,w0)+b0) # 첫번째 회귀
    hypothesis1 = tf.nn.softmax(tf.matmul(hypothesis0,w1)+b1) # 두번째 회귀
    print("hypothesis1=",hypothesis1[0:3])
    print("y_train=",y_train[0:3])
    cost = -tf.reduce_sum(y_train * tf.log(hypothesis1)) / len(hypothesis1)
    print("cost=",cost) 
    grads = tape.gradient(cost, [w0,w1,b0,b1])
    # # print("grads=",grads)
    # print("w0 업제이트 전=",w0)
    # print("w1 업제이트 전=",w1)
    # print("b0 업제이트 전=",b0)
    # print("b1 업제이트 전=",b1)
    optimizer.apply_gradients(grads_and_vars=zip(grads,[w0,w1,b0,b1]))
    # print("w0 업제이트 후=",w0)
    # print("w1 업제이트 후=",w1)
    # print("b0 업제이트 후=",b0)
    # print("b1 업제이트 후=",b1)
    print("="*100)
```
