```python
fig = plt.figure(figsize=(15,10))
for i in range(12):
    subplot = fig.add_subplot(3,4,i+1)
    subplot.set_yticks([])
    subplot.set_title("label : %d" % y[i])
    subplot.imshow(X[i].reshape(28,28), cmap=plt.cm.gray_r)
```
