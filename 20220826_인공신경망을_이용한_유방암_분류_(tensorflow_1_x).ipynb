```python
X = np.array(X, dtype="float32") # X의 데이터 타입을 float32로 변환 
X
```

```python
# 양성 (pasitive) = 찾는 거 = 1
# 악성 (negative) = 관심 없음 = 0

y = cancer.target  # 종속변수
y                  # 악성이면 0 양성이면 1
```

```python
y = np.array(y, dtype="float32") # y의 데이터 타입을 float32로 변환 
y
```

```python
y = y.reshape(1,-1) # 1줄 칸(생략)으로 변환
y
```

```python
# y.reshape() = 1차원 배열을 2차원 배열로 변환
# y.reshape(1,-1) = 행 형태의 2차원 배열 ( -1은 생략이라는 뜻 )
# y.reshape(-1,1) = 열 형태의 2차원 배열 ( -1은 생략이라는 뜻 )
# 현재 X가 2차원 배열이기 때문에 y도 변환
# y = y.reshape(569,1) # 569줄 1칸으로 변환

y = y.reshape(-1,1) # 줄(생략) 1칸으로 변환
y
```

```python
from tensorflow_core import OptimizerOptions
optimizer = tf.train.AdamOptimizer(0.001)
```

```python
with tf.GradientTape() as tape:

    hypothesis = tf.sigmoid( tf.matmul(X_train , w) + b )
    print("hypothesis = ", hypothesis[0:3])

    cost = -tf.reduce_mean( (y_train*tf.log(hypothesis)\
            +(1-y_train)*tf.log(1-hypothesis)) ) 
    print("cost = ", cost)

    grads = tape.gradient(cost, [w,b])
    print("기울기 grads = ", grads)
    
    print("업데이트 전 w =", w)
    print("업데이트 전 b =", b)
    optimizer.apply_gradients(
        grads_and_vars = zip(grads, [w, b])
    )
    print("업데이트 후 w =", w)
    print("업데이트 후 b =", b)
    print("="*100)
```

```python
for step in range(100):
    print("step =", step)
    
    with tf.GradientTape() as tape:

        hypothesis = tf.sigmoid( tf.matmul(X_train , w) + b )
        print("hypothesis = ", hypothesis[0:3])

        cost = -tf.reduce_mean( (y_train*tf.log(hypothesis)\
                +(1-y_train)*tf.log(1-hypothesis)) ) 
        print("cost = ", cost)

        grads = tape.gradient(cost, [w,b])
        print("기울기 grads = ", grads)
        
        print("업데이트 전 w =", w)
        print("업데이트 전 b =", b)
        optimizer.apply_gradients(
            grads_and_vars = zip(grads, [w, b])
        )
        print("업데이트 후 w =", w)
        print("업데이트 후 b =", b)
        print("="*100)
```
