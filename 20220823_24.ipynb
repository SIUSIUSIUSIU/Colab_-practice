```python
import numpy as np 
```

```python
np.array([1,2,3], dtype="float32")
```

```python
X = np.array([1,2,3], dtype="float32")
```

```python
X
```

```python
y = np.array([2, 2.5, 3.5], dtype="float32")
```

```python
y
```

```python
# w, b 값 랜덤
w = 2
b = 1
```

```python
# hypothesis : 예측값
hypothesis = X*w+b
hypothesis
```

```python
# (예측값 - 실제값) 제곱
(hypothesis - y) ** 2
```

```python
# (예측값 - 실제값) 제곱 합 / 데이터 수
cost = sum( (hypothesis - y) ** 2) / 3
```

```python
# cost : 오차값
cost
```

```python
# 1/3*np.dot((hypothesis-y)) # 1/m * ( 예측값 - 실제값 )
```

```python
w_gred = 1/3*np.dot((hypothesis-y) , X.T)
```

```python
w_gred #기울기
```

```python
w = w - w_gred
w
```

```python
b_gred = 1/3*np.sum(hypothesis-y)
b_gred #기울기
```

```python
b = b - b_gred
b
```

```python
w = 2
b = 1
learning_rate=0.1

for i in range(1000):
    hypothesis = X*w+b
    print("hypothesis:", hypothesis)
    MES=np.sum((hypothesis-y)**2)
    print("MES:", MES)
    w_gred = 1/3*np.dot((hypothesis-y) , X.T)
    print("w_gred:", w_gred)
    b_gred = 1/3*np.sum(hypothesis-y)
    print("b_gred:", b_gred)
    w = w - (learning_rate * w_gred)
    print("w:", w)
    b = b - (learning_rate * b_gred)
    print("b:", b)
```
